Which vendors and brands demonstrate highest sales performance?

```python
"""
This block identifies the top 10 vendors and top 10 product brands 
based on their total sales revenue. It helps highlight the 
highest-performing contributors to overall sales.
"""

# Define a helper function to format large dollar values (like 1.2M or 850K)
def format_dollars(value):
    if value >= 1_000_000:
        return f"{value / 1_000_000:.2f}M"   # Convert to millions if >= 1 million
    elif value >= 1_000:
        return f"{value / 1_000:.2f}K"       # Convert to thousands if >= 1,000
    else:
        return f"{value:.2f}"                # Keep original value if small

# Group the data by VendorName and sum their TotalSalesDollars
top_vendors = df.groupby("VendorName")["TotalSalesDollars"].sum().nlargest(10)
#  Returns top 10 vendors by total sales revenue

# Group the data by product Description and sum their TotalSalesDollars
top_brands = df.groupby("Description")["TotalSalesDollars"].sum().nlargest(10)
#  Returns top 10 selling product brands by revenue

print(top_vendors.apply(lambda x :format_dollars(x)))# Display the top 10 vendors
print()
print(top_brands.apply(lambda x :format_dollars(x)))  # Display the top 10 brands
```

```python
"""
This code creates side-by-side bar plots for visualizing the Top 10 Vendors and Top 10 Brands 
based on total sales. It also adds sales value labels beside each bar for better interpretation.
"""

plt.figure(figsize=(15,5))  # Create a wide figure for two side-by-side subplots

#  Plot for Top Vendors
plt.subplot(1,2,1)  # First subplot in a 1-row, 2-column grid
ax1 = sns.barplot(y=top_vendors.index, x=top_vendors.values, palette="Blues_r")  
#  Horizontal bar plot for top vendors using blue color palette
plt.title("Top 10 Vendors by Sales")  # Add title

#  Add sales value labels to each bar
for bar in ax1.patches:
    ax1.text(
        bar.get_width() + (bar.get_width() * 0.02),  # X-position (slightly beyond bar width)
        bar.get_y() + bar.get_height() / 2,          # Y-position (middle of bar)
        format_dollars(bar.get_width()),             # Convert width (sales) to $ format
        ha='left', va='center', fontsize=10, color='black'  # Text formatting
    )

#  Plot for Top Brands
plt.subplot(1,2,2)  # Second subplot
ax2 = sns.barplot(y=top_brands.index.astype(str), x=top_brands.values, palette="Reds_r")
#  Horizontal bar plot for top brands using red color palette
plt.title("Top 10 Brands by Sales")  # Add title

#  Add sales value labels to each bar
for bar in ax2.patches:
    ax2.text(
        bar.get_width() + (bar.get_width() * 0.02), 
        bar.get_y() + bar.get_height() / 2, 
        format_dollars(bar.get_width()), 
        ha='left', va='center', fontsize=10, color='black'
    )

plt.tight_layout()  # Adjust subplot spacing to avoid overlap
plt.show()          # Display both bar plots
```

Which Vendors contribute the most to total purchase dollars?

```python
"""
This block analyzes vendor-level contribution to total purchases, sales, and profit. 
It helps identify top-performing vendors by calculating their share of total purchase 
value and ranks the top 10 vendors accordingly.
"""

# Group data by VendorName and aggregate total purchases, sales, and gross profit
vendor_performance = df.groupby('VendorName').agg({
    'TotalPurchaseDollars': 'sum',    # Sum of purchase value per vendor
    'GrossProfit': 'sum',             # Sum of gross profit per vendor
    'TotalSalesDollars': 'sum'        # Sum of sales revenue per vendor
}).reset_index()                      # Reset index to keep VendorName as a column

# Calculate each vendor's contribution to total purchases (in percentage)
vendor_performance['PurchaseContribution%'] = (
    vendor_performance['TotalPurchaseDollars'] / vendor_performance['TotalPurchaseDollars'].sum()*100
)

# Round values and sort vendors in descending order of purchase contribution
vendor_performance = round(vendor_performance.sort_values(by='PurchaseContribution%', ascending=False), 2)

# Select top 10 vendors contributing the most to total purchases
top_vendors = vendor_performance.head(10)

# Format dollar values for better readability (assuming format_dollars is a custom formatting function)
top_vendors['TotalSalesDollars'] = top_vendors['TotalSalesDollars'].apply(format_dollars)
top_vendors['TotalPurchaseDollars'] = top_vendors['TotalPurchaseDollars'].apply(format_dollars)
top_vendors['GrossProfit'] = top_vendors['GrossProfit'].apply(format_dollars)

# Display the top 10 vendors with formatted values
top_vendors

# Print total number of vendors and total contribution of top 10 vendors
print('Total Vendors', vendor_performance.shape)  #  Shows count of unique vendors
print("Purchase contribution of top 10 vendors", top_vendors['PurchaseContribution%'].sum())
#  Helps understand what % of purchase value is concentrated among top 10 vendors

# Calculate cumulative contribution of top vendors to visualize Pareto trend
top_vendors['Cumulative_Contribution%'] = top_vendors['PurchaseContribution%'].cumsum()
#  Adds up each vendors contribution to see combined impact

# Display final summary table
top_vendors
```

```python
"""
This code creates a Pareto Chart to analyze vendor-wise contribution 
to total purchases using both bar and line plots. It shows which vendors 
account for the majority of the purchase volume.
"""

fig, ax1 = plt.subplots(figsize=(12, 6))  # Initialize plot and primary axis

# Bar Plot for individual vendor purchase contribution
sns.barplot(
    x=top_vendors['VendorName'], 
    y=top_vendors['PurchaseContribution%'], 
    palette="mako", 
    ax=ax1
)

# Add text labels on top of each bar
for i, value in enumerate(top_vendors['PurchaseContribution%']):
    ax1.text(i, value - 1, str(value) + '%', ha='center', fontsize=10, color='white')

# Create a secondary y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Line Plot for Cumulative Contribution %
ax2.plot(
    top_vendors['VendorName'], 
    top_vendors['Cumulative_Contribution%'], 
    color='red', 
    marker='o', 
    linestyle='dashed', 
    label='Cumulative%'
)

# Formatting X and Y axes
ax1.set_xticklabels(top_vendors['VendorName'], rotation=90)              # Rotate vendor names for readability
ax1.set_ylabel('Purchase Contribution %', color='blue')                  # Y-axis label for bar chart
ax2.set_ylabel('Cumulative Contribution %', color='red')                 # Y-axis label for line chart
ax1.set_xlabel('Vendors')                                                # X-axis label
ax2.set_title('Pareto Chart: Vendor Contribution to Total Purchases')    # Chart title

# Add horizontal line at 100% as visual reference
ax2.axhline(y=100, color='gray', linestyle='dashed', alpha=0.7)

# Show legend for line plot
ax2.legend(loc='upper right')

# Display the final plot
plt.tight_layout()
plt.show()
```

What is the 95% confidence interval for profit margins of top performing and low performing vendors(We want to estimate the typical profit margin range (with 95% certainty) for vendors who perform the best and the worst.)

```python
"""
This code compares the Profit Margin distributions of top vs. low-performing vendors 
(based on Total Sales Dollars) and calculates their 95% Confidence Intervals (CI). 
It visualizes these insights using histograms and confidence bands.
"""

# Define thresholds for top 25% and bottom 25% vendors based on TotalSalesDollars
top_threshold = df["TotalSalesDollars"].quantile(0.75)  # 75th percentile value
low_threshold = df["TotalSalesDollars"].quantile(0.25)  # 25th percentile value

# Get ProfitMargin data for top and low vendors
top_vendors = df[df['TotalSalesDollars'] >= top_threshold]["ProfitMargin"].dropna()
low_vendors = df[df['TotalSalesDollars'] <= low_threshold]["ProfitMargin"].dropna()

# Print values (optional for inspection)
print(top_vendors)
print()
print(low_vendors)

# Define function to compute 95% confidence interval
def confidence_interval(data, confidence=0.95):
    mean_val = np.mean(data)                                      # Mean of the data
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))           # Standard error
    t_critical = stats.t.ppf((1 + confidence) / 2, df=len(data)-1) # t-score for CI
    margin_of_error = t_critical * std_err                        # Margin of error
    return mean_val, mean_val - margin_of_error, mean_val + margin_of_error

# Calculate CI for top and low vendors
top_mean, top_lower, top_upper = confidence_interval(top_vendors)
low_mean, low_lower, low_upper = confidence_interval(low_vendors)

# Print CI results
print(f"Top Vendors 95% CI: ({top_lower:.2f}, {top_upper: .2f}), Mean: {top_mean: .2f}")
print(f"Low Vendors 95% CI: ({low_lower:.2f}, {low_upper:.2f}), Mean: {low_mean: .2f}")

plt.figure(figsize=(12, 6))  # Set size of the plot

# Plot histogram and KDE for Top Vendors
sns.histplot(top_vendors, kde=True, color='blue', bins=30, alpha=0.5, label="Top Vendors")
plt.axvline(top_lower, color="blue", linestyle='--', label=f"Top Lower: {top_lower:.2f}")
plt.axvline(top_upper, color="blue", linestyle='--', label=f"Top Upper: {top_upper:.2f}")
plt.axvline(top_mean,  color="blue", linestyle='--', label=f"Top Mean: {top_mean:.2f}")

# Plot histogram and KDE for Low Vendors
sns.histplot(low_vendors, kde=True, color='red', bins=30, alpha=0.5, label="Low Vendors")
plt.axvline(low_lower, color="red", linestyle='--', label=f"Low Lower: {low_lower:.2f}")
plt.axvline(low_upper, color="red", linestyle='--', label=f"Low Upper: {low_upper:.2f}")
plt.axvline(low_mean,  color="red", linestyle='--', label=f"Low Mean: {low_mean:.2f}")

# Finalize plot
plt.title("Confidence Interval Comparison: Top vs. Low Vendors (Profit Margin)") 
plt.xlabel("Profit Margin (%)")       # X-axis label
plt.ylabel("Frequency")               # Y-axis label
plt.legend()                          # Show legend
plt.grid(True)                        # Add grid for readability
plt.show()                            # Display the plot
```

* The confidence interval for low-performing vendors (40.48% to 42.62%) is signifidantly higher than that of top-performing vendors (30.74% to 31.61%).
* This suggests that vendors with lower sales tend to maintain higher profit margins, potentially due to premium pricing or lower operational costs.
* For High-Performing Vendors: If they aim to improve profitability, they could explore selective price adjustments, cost optimization, or bundling strategies.
* For Low-Performing Vendors: Despite higher margins, their low sales volume might indicate a need for better marketing, competitive pricing, or improved distribution strategies.

### Is there a significant difference in profit margins between top-performing and low-performing vendors?
Hypothesis:
Ho (Null Hypothesis): There is no significant difference in the mean profit margins of top-performing and low-erforming vendors.<br><br>
H1 (Alternative Hypothesis): The mean profit margins of top-performing and low-performing vendors are significantly different.

```python
top_threshold = df["TotalSalesDollars"].quantile(0.75)  # 75th percentile value
low_threshold = df["TotalSalesDollars"].quantile(0.25)  # 25th percentile value

top_vendors = df[df['TotalSalesDollars'] >= top_threshold]["ProfitMargin"].dropna()
low_vendors = df[df['TotalSalesDollars'] <= low_threshold]["ProfitMargin"].dropna()

t_stat,p_value = ttest_ind(top_vendors,low_vendors,equal_var=False)

#Print results
print(f"T-statistic:{t_stat:.4f},P-value:{p_value:.4f}")
if p_value <0.05:
    print("Reject H0: There is a significant difference in profit margins between top and low performing vendors.")
else:
    print("Failed to Reject H0: No significant difference in profit margin")
    
```
